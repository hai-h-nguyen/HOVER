```python
import numpy as np
```

```python
play = np.load("/home/rtx4/HOVER/play.npy")[0]
play
```

```python
sim2sim = np.load("/home/rtx4/HOVER/sim2sim.npy")[0]
sim2sim
```

```python
diff = play - sim2sim
diff
```

```python
import matplotlib.pyplot as plt

plt.plot(diff[:200], label='Difference (A - B)', color='r')
plt.title('Difference between corresponding elements of A and B')
plt.xlabel('Index')
plt.ylabel('Difference')
plt.legend()
plt.show()
```

```python
pos = np.where(diff != 0)[0]
pos
```

```python
print(pos.shape)
```

```python
print(play[pos])
```

```python
print(sim2sim[pos])
```
